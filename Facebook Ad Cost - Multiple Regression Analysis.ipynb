```python
#import and clean up the data for analysis
#data is available on GitHub
#------------------------------------------------#

#import the pandas package into script
import pandas

#Import Travel Pony csv as a dataframe
trvlp = 'Travel Pony Facebook.csv'

#travel pony data as a data frame
trvlp = pandas.read_csv('Travel Pony Facebook.csv')

#remove any rows that have $0 spend - we do not need these.
trvl = trvlp[trvlp['Amount Spent (USD)'] != 0]
#------------------------------------------------#
```

```python
#import seaborn package for analysis
#------------------------------------------------#

#import seaborn as sns
import seaborn as sns

#Question / Step 1
#Graph histogram of amount spent
sns.set_style('whitegrid')  #makes the background of the grid white
sns.distplot(trvl['Amount Spent (USD)']) #creates the distribution plot
#------------------------------------------------#
```

```python
#------------------------------------------------#
#preparing the data to fit a multiple regression model
#multiple column approach
#------------------------------------------------#


#assign categorical numbers to the campaign names
trvl['Campaign Name'] = trvl['Campaign Name'].astype('category')
trvl['Campaign Code'] = trvl['Campaign Name'].cat.codes #this gives the categories codes

#add the (multiple) predictor variables
X = trvl[['Reach', 'Frequency', 'Impressions', 'Clicks', 'Unique Clicks', 'Page Likes','Page Engagement', 'Post Engagement', 'Post Likes', 'Post Comments','Post Shares', 'Photo Views', 'Website Clicks', 'Campaign Code']]

#import statsmodels.api as sm
import statsmodels.api as sm

#create y 
y = trvl['Amount Spent (USD)']

linreg = sm.OLS(y,X).fit()

linreg.summary()
#------------------------------------------------#
```
